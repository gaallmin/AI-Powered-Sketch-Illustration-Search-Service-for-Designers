```python
from PIL import Image
import matplotlib.pyplot as plt
import os
import random
```

```python
L2_3_root_path = "../dataset/L2_3/"
L2_3_name_list = os.listdir("../dataset/L2_3")
L2_10_root_path = "../dataset/L2_10/"
L2_10_name_list = os.listdir("../dataset/L2_10")
```

```python
L2_3_random_file_list = random.choices(L2_3_name_list, k=20)
L2_10_random_file_list = random.choices(L2_10_name_list, k=20)

fig, axes = plt.subplots(8, 5)
fig.set_size_inches(10.5, 15.5, forward=True)

for file_idx, file_name in enumerate(L2_3_random_file_list):
    image = Image.open(L2_3_root_path + file_name)
    
    col = file_idx % 5
    row = file_idx // 5
    axes[row][col].imshow(image)
    
for file_idx, file_name in enumerate(L2_10_random_file_list):
    image = Image.open(L2_10_root_path + file_name)
    
    col = file_idx % 5
    row = 4 + file_idx // 5
    axes[row][col].imshow(image)
```
